Q1. What is hypothesis testing in statistics?

--->Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It involves:

Making an assumption (called the null hypothesis, H‚ÇÄ)

Testing it against an alternative hypothesis (H‚ÇÅ)

Using statistical evidence to accept or reject H‚ÇÄ

The goal is to determine whether the observed data is due to chance or if there is enough evidence to support a different claim.

Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

--->The null hypothesis (H‚ÇÄ) is a statement that assumes there is no effect or no difference in a situation or experiment. It represents the status quo or a default assumption.

The alternative hypothesis (H‚ÇÅ or Ha) is the statement that contradicts the null hypothesis. It suggests that there is an effect or a difference.

Q3. What is the significance level in hypothesis testing, and why is it important?

--->The significance level in hypothesis testing, denoted by Œ± (alpha), is the probability of rejecting the null hypothesis (H‚ÇÄ) when it is actually true. It represents the risk of a Type I error.

Common Values: 0.05 (5%)

0.01 (1%)

Importance: It sets the threshold for decision-making.

Helps determine whether the test result is statistically significant.

A lower Œ± means stricter criteria, reducing the chance of a false positive.

Example: If Œ± = 0.05, there's a 5% chance you'll incorrectly reject H‚ÇÄ when it's actually true.

Q4. What does a P-value represent in hypothesis testing?

--->A P-value in hypothesis testing represents the probability of observing the test results, or something more extreme, assuming the null hypothesis (H‚ÇÄ) is true.

In Simple Terms: It tells you how likely your sample result is due to chance.

A small P-value (typically < 0.05) suggests strong evidence against H‚ÇÄ, so you reject H‚ÇÄ.

A large P-value means the result is likely under H‚ÇÄ, so you fail to reject H‚ÇÄ.

Example: If P-value = 0.03, there's a 3% chance the observed result happened by random chance under H‚ÇÄ. This would be considered statistically significant at the 0.05 level.

Q5. How do you interpret the P-value in hypothesis testing?

--->To interpret the P-value in hypothesis testing:

Compare it with the significance level (Œ±), usually 0.05.

Interpretation: If P-value ‚â§ Œ±: ‚Üí Reject the null hypothesis (H‚ÇÄ) ‚Üí The result is statistically significant ‚Üí There is evidence for the alternative hypothesis (H‚ÇÅ)

If P-value > Œ±: ‚Üí Fail to reject the null hypothesis (H‚ÇÄ) ‚Üí The result is not statistically significant ‚Üí Not enough evidence to support H‚ÇÅ

Example: P-value = 0.02, Œ± = 0.05 ‚Üí 0.02 ‚â§ 0.05 ‚Üí Reject H‚ÇÄ ‚Üí Significant result

P-value = 0.08, Œ± = 0.05 ‚Üí 0.08 > 0.05 ‚Üí Fail to reject H‚ÇÄ ‚Üí Not significant

Q6. What are Type 1 and Type 2 errors in hypothesis testing?

Type I Error (False Positive): Meaning: Rejecting the null hypothesis (H‚ÇÄ) when it is actually true

Probability: Denoted by Œ± (significance level)

Type II Error (False Negative): Meaning: Failing to reject the null hypothesis (H‚ÇÄ) when it is actually false

Example: Concluding a drug doesn't work when it actually does

Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy import stats

# Sample data
sample_data = [72, 68, 75, 70, 74, 71, 69, 73, 72, 76]

# Given population parameters
population_mean = 70
population_std = 3  # Assume known population standard deviation

# Step 1: Calculate sample mean and size
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Step 2: Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Step 3: Calculate p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Step 4: Set significance level
alpha = 0.05

# Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-score = {z_score:.4f}")
print(f"P-value = {p_value:.4f}")

# Step 5: Interpret results
if p_value < alpha:
    print("Result: Reject the null hypothesis (statistically significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
```

Q4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Sample data
sample_data = [102, 98, 105, 101, 99, 100, 104, 97, 103, 100]

# Known population parameters
population_mean = 100
population_std = 5  # Known œÉ
n = len(sample_data)
sample_mean = np.mean(sample_data)

# Perform Z-test
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Set significance level
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha / 2)  # For two-tailed test

# Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-score = {z_score:.4f}")
print(f"P-value = {p_value:.4f}")
if abs(z_score) > z_critical:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")

# -------- Plotting the decision regions --------
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution', color='black')

# Shade rejection regions
plt.fill_between(x, y, where=(x < -z_critical), color='red', alpha=0.3, label='Rejection Region (left)')
plt.fill_between(x, y, where=(x > z_critical), color='red', alpha=0.3, label='Rejection Region (right)')

# Plot Z-score
plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score = {z_score:.2f}')

# Mark critical values
plt.axvline(-z_critical, color='grey', linestyle='dotted', label=f'Critical Z = ¬±{z_critical:.2f}')
plt.axvline(z_critical, color='grey', linestyle='dotted')

# Labels and legend
plt.title("Two-Tailed Z-Test Decision Regions")
plt.xlabel("Z-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05, two_tailed=True):
    """
    Calculate and visualise Type‚ÄØI (Œ±) and Type‚ÄØII (Œ≤) errors for a Z‚Äëtest on the mean.

    Parameters
    ----------
    mu0 : float
        Mean under the null hypothesis.
    mu1 : float
        True (actual) mean under the alternative hypothesis.
    sigma : float
        Known population standard deviation.
    n : int
        Sample size.
    alpha : float, default 0.05
        Significance level.
    two_tailed : bool, default True
        If True perform a two‚Äëtailed test, else right‚Äëtailed (greater‚Äëthan) test.
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value(s) and region limits
    if two_tailed:
        z_crit = stats.norm.ppf(1 - alpha / 2)
        lower_crit = mu0 - z_crit * se
        upper_crit = mu0 + z_crit * se
    else:
        z_crit = stats.norm.ppf(1 - alpha)
        lower_crit = -np.inf
        upper_crit = mu0 + z_crit * se  # right‚Äëtailed test

    # Type‚ÄØI error (by construction equals alpha for two‚Äëtailed)

     type1 = alpha

    # Type‚ÄØII error (probability of failing to reject H0 when H1 is true)
    beta = (stats.norm.cdf(upper_crit, loc=mu1, scale=se) -
            stats.norm.cdf(lower_crit, loc=mu1, scale=se))
    power = 1 - beta

    # ------- Visualisation -------
    x_min = min(mu0, mu1) - 4*se
    x_max = max(mu0, mu1) + 4*se
    x = np.linspace(x_min, x_max, 1000)

    null_pdf = stats.norm.pdf(x, loc=mu0, scale=se)
    alt_pdf = stats.norm.pdf(x, loc=mu1, scale=se)

    plt.figure(figsize=(10, 5))

    # Plot the two distributions
    plt.plot(x, null_pdf, label='Null Distribution (H0)')
    plt.plot(x, alt_pdf, label='Alternative Distribution (H1)', linestyle='--')

    # Shade Type‚ÄØI rejection regions (under H0)
    if two_tailed:
        plt.fill_between(x, 0, null_pdf, where=(x <= lower_crit), alpha=0.3, label='Type‚ÄØI error region')
        plt.fill_between(x, 0, null_pdf, where=(x >= upper_crit), alpha=0.3)
    else:  # right‚Äëtailed
        plt.fill_between(x, 0, null_pdf, where=(x >= upper_crit), alpha=0.3, label='Type‚ÄØI error region')

    # Shade Type‚ÄØII region (under H1)
    plt.fill_between(x, 0, alt_pdf, where=((x >= lower_crit) & (x <= upper_crit)), alpha=0.3, label='Type‚ÄØII error region')

    # Critical lines
    plt.axvline(lower_crit, linestyle='dotted')
    plt.axvline(upper_crit, linestyle='dotted')

    # Labels
    plt.title('Type‚ÄØI (Œ±) and Type‚ÄØII (Œ≤) Errors')
    plt.xlabel('Sample mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print results
    print(f"Significance level (Œ±)  : {type1:.3f}")
    print(f"Type‚ÄØII error (Œ≤)       : {beta:.3f}")
    print(f"Statistical Power (1-Œ≤) : {power:.3f}")

# --- Example run ---
type1_type2_errors(mu0=0, mu1=1, sigma=10, n=30, alpha=0.05, two_tailed=True)
```

Q6.Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy import stats

# -----------------------------
# 1. Sample data (replace with your own)
# -----------------------------
group_A = np.array([12, 15, 14, 10, 13, 15, 14, 16, 11, 13])
group_B = np.array([18, 20, 17, 19, 22, 21, 20, 18, 17, 19])

# -----------------------------
# 2. Independent two‚Äësample t‚Äëtest (Welch‚Äôs t)
# -----------------------------
t_stat, p_val = stats.ttest_ind(group_A, group_B, equal_var=False)

# -----------------------------
# 3. Output and interpretation
# -----------------------------
alpha = 0.05            # significance level
print(f"Group A mean = {group_A.mean():.2f}")
print(f"Group B mean = {group_B.mean():.2f}")
print(f"t‚Äëstatistic  = {t_stat:.4f}")
print(f"p‚Äëvalue      = {p_val:.4f}")

if p_val < alpha:
    print(f"Result: Reject H‚ÇÄ (means differ at Œ± = {alpha})")
else:
    print(f"Result: Fail to reject H‚ÇÄ (no significant difference at Œ± = {alpha})")
```

Q7. Perform a paired sample T-test using Python and visualize the comparison results.

Q9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy import stats

def mean_confidence_interval(sample, confidence=0.95, pop_std=None):
    """
    Calculate a (1‚ÄëŒ±)% confidence interval for the population mean Œº.

    Parameters
    ----------
    sample      : array‚Äëlike of numbers
    confidence  : float, desired confidence level (default 0.95)
    pop_std     : float or None. If None, œÉ is unknown and a t‚Äëinterval is used;
                  otherwise a z‚Äëinterval is computed with the supplied œÉ.

    Returns
    -------
    (lower, upper) tuple representing the confidence interval.
    """
    sample = np.asarray(sample, dtype=float)
    n       = sample.size
    mean    = sample.mean()
    alpha   = 1 - confidence

    if pop_std is None:                               # --- t‚Äëinterval (œÉ unknown) ---
        s      = sample.std(ddof=1)
        se     = s / np.sqrt(n)                       # standard error
        df     = n - 1
        crit   = stats.t.ppf(1 - alpha / 2, df)       # critical t
    else:                                             # --- z‚Äëinterval (œÉ known) ---
        se     = pop_std / np.sqrt(n)
        crit   = stats.norm.ppf(1 - alpha / 2)        # critical z

    margin = crit * se
    return mean - margin, mean + margin
np.random.seed(42)
data = np.random.normal(loc=75, scale=8, size=40)

ci_95 = mean_confidence_interval(data, confidence=0.95)        # œÉ unknown
ci_90 = mean_confidence_interval(data, confidence=0.90, pop_std=8)  # œÉ known

print("95% CI (t) :", ci_95)
print("90% CI (z) :", ci_90)
```

It conveys both the estimate (its centre) and the uncertainty (its width).

Narrow CI ‚Üí high precision (large‚ÄØn, low variability).

Wide‚ÄØCI ‚Üí greater uncertainty (small‚ÄØn, high variability).

In hypothesis testing, if a 95‚ÄØ% CI excludes the null value (e.g., a mean difference of‚ÄØ0), you would reject the null at the 0.05 level‚Äîso CIs give the same information as a two‚Äëtailed significance test, but with more nuance.

Q10.Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy import stats

def margin_of_error(sample, confidence=0.95, pop_std=None):
    """
    Calculate the margin of error for the sample mean at a given confidence level.

    Parameters:
    - sample: list or array of sample data
    - confidence: confidence level (default = 0.95)
    - pop_std: known population standard deviation (if any)

    Returns:
    - Margin of Error (float)
    """
    sample = np.asarray(sample)
    n = len(sample)
    alpha = 1 - confidence

    if pop_std is None:
        # Use t-distribution (population std unknown)
        sample_std = np.std(sample, ddof=1)
        se = sample_std / np.sqrt(n)
        critical_value = stats.t.ppf(1 - alpha / 2, df=n - 1)
    else:
        # Use z-distribution (population std known)
        se = pop_std / np.sqrt(n)
        critical_value = stats.norm.ppf(1 - alpha / 2)

    moe = critical_value * se
    return moe
```

Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(p_disease, p_pos_given_disease, p_neg_given_no_disease):
    # Step 1: Convert specificity to false positive rate
    p_pos_given_no_disease = 1 - p_neg_given_no_disease

    # Step 2: Compute prior probabilities
    p_no_disease = 1 - p_disease

    # Step 3: Total probability of testing positive (law of total probability)
    p_positive = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)

    # Step 4: Apply Bayes' Theorem
    p_disease_given_pos = (p_pos_given_disease * p_disease) / p_positive

    return p_disease_given_pos

# Input probabilities
p_disease = 0.01              # 1% prevalence
p_pos_given_disease = 0.99    # 99% sensitivity
p_neg_given_no_disease = 0.95 # 95% specificity

# Run the inference
posterior = bayes_theorem(p_disease, p_pos_given_disease, p_neg_given_no_disease)
print(f"Probability of having the disease given a positive test: {posterior:.4f} ({posterior*100:.2f}%)")
```

Probability of having the disease given a positive test: 0.1667 (16.67%)
Explanation of the Process Prior (belief before seeing data)

Probability of having the disease: P(Disease) = 0.01

Likelihood (how likely the evidence is)

P(Positive | Disease) = 0.99

Marginal probability (overall probability of testing positive)

Combines true and false positives from both groups.

Posterior (updated belief after evidence)

P(Disease | Positive) ‚Äî what we really want!

Q16.Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy.stats import f

# -----------------------------
# 1. Simulate two random samples
# -----------------------------
np.random.seed(42)

sample1 = np.random.normal(loc=50, scale=5, size=30)  # Sample 1 with variance ‚âà 25
sample2 = np.random.normal(loc=50, scale=10, size=30) # Sample 2 with variance ‚âà 100

# -----------------------------
# 2. Compute sample variances
# -----------------------------
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# -----------------------------
# 3. Compute F-statistic
# -----------------------------
if var1 > var2:
    f_stat = var1 / var2
    dfn, dfd = len(sample1) - 1, len(sample2) - 1
else:
    f_stat = var2 / var1
    dfn, dfd = len(sample2) - 1, len(sample1) - 1

# -----------------------------
# 4. Calculate the two-tailed p-value
# -----------------------------
p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

# -----------------------------
# 5. Output results
# -----------------------------
{
    "Sample 1 Variance": round(var1, 3),
    "Sample 2 Variance": round(var2, 3),
    "F-statistic": round(f_stat, 4),
    "Degrees of Freedom": (dfn, dfd),
    "P-value": round(p_value, 4),
    "Conclusion (Œ± = 0.05)": "Reject H‚ÇÄ (variances differ)" if p_value < 0.05 else "Fail to reject H‚ÇÄ (no significant difference)"
}
```

Q19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, levene
import statsmodels.api as sm
import statsmodels.formula.api as smf

def check_anova_assumptions(data: pd.DataFrame, group_col: str, value_col: str, plot=True):
    """
    Check normality and homogeneity of variance for one-way ANOVA.

    Parameters:
    - data: DataFrame with the data
    - group_col: column name for group labels
    - value_col: column name for measured values
    - plot: whether to show Q-Q plot of residuals

    Returns:
    - Dictionary with Shapiro-Wilk p-value (normality) and Levene's test p-value (equal variances)
    """

    # Fit one-way ANOVA model
    model = smf.ols(f"{value_col} ~ C({group_col})", data=data).fit()
    residuals = model.resid

    # 1. Test normality of residuals (Shapiro-Wilk)
    shapiro_stat, shapiro_p = shapiro(residuals)

    # 2. Test homogeneity of variances (Levene's test)
    grouped_data = [group[value_col].values for _, group in data.groupby(group_col)]
    levene_stat, levene_p = levene(*grouped_data)

    if plot:
        sm.qqplot(residuals, line='s')
        plt.title("Q-Q Plot of Residuals")
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.show()

    return {
        "Shapiro-Wilk p-value (normality)": round(shapiro_p, 4),
        "Levene‚Äôs test p-value (equal variance)": round(levene_p, 4),
        "Conclusion (Œ± = 0.05)": {
            "Normality": "Pass" if shapiro_p > 0.05 else "Fail",
            "Equal Variance": "Pass" if levene_p > 0.05 else "Fail"
        }
    }
```

Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
from scipy.stats import chi2

# -----------------------------
# 1. Simulate sample data
# -----------------------------
np.random.seed(0)
sample = np.random.normal(loc=100, scale=10, size=30)  # true std = 10, variance = 100

# -----------------------------
# 2. Hypothesized population variance (œÉ¬≤‚ÇÄ)
# -----------------------------
sigma_squared_0 = 100  # claim: population variance is 100

# -----------------------------
# 3. Compute test statistic
# -----------------------------
n = len(sample)
sample_variance = np.var(sample, ddof=1)
chi2_stat = (n - 1) * sample_variance / sigma_squared_0

# -----------------------------
# 4. Calculate p-value for two-tailed test
# -----------------------------
alpha = 0.05
df = n - 1
p_value = 2 * min(chi2.cdf(chi2_stat, df), 1 - chi2.cdf(chi2_stat, df))

# -----------------------------
# 5. Print results
# -----------------------------
print(f"Sample Variance     : {sample_variance:.4f}")
print(f"Hypothesized Var œÉ¬≤‚ÇÄ: {sigma_squared_0}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"Degrees of Freedom  : {df}")
print(f"P-value             : {p_value:.4f}")
# Decision
if p_value < alpha:
    print("Conclusion: Reject H‚ÇÄ ‚Üí Sample variance differs from hypothesized variance.")
else:
    print("Conclusion: Fail to reject H‚ÇÄ ‚Üí No significant difference in variance.")
```

Q25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# -----------------------------
# 1. Define successes and sample sizes
# -----------------------------
successes = np.array([60, 80])      # Number of "successes" in each group
nobs = np.array([200, 250])         # Total observations per group

# -----------------------------
# 2. Perform two-proportion Z-test
# -----------------------------
z_stat, p_value = proportions_ztest(count=successes, nobs=nobs, alternative='two-sided')

# -----------------------------
# 3. Output the results
# -----------------------------
print(f"Proportion A: {successes[0] / nobs[0]:.4f}")
print(f"Proportion B: {successes[1] / nobs[1]:.4f}")
print(f"Z-statistic : {z_stat:.4f}")
print(f"P-value     : {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H‚ÇÄ ‚Üí Proportions are significantly different.")
else:
    print("Conclusion: Fail to reject H‚ÇÄ ‚Üí No significant difference in proportions.")
```

Q26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f, f_oneway

# -----------------------------
# 1. Simulate two datasets with different variances
# -----------------------------
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)   # Variance ‚âà 25
group2 = np.random.normal(loc=50, scale=10, size=30)  # Variance ‚âà 100

# -----------------------------
# 2. Compute sample variances
# -----------------------------
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

# Ensure the larger variance is in the numerator
if var1 > var2:
    f_stat = var1 / var2
    dfn, dfd = len(group1) - 1, len(group2) - 1
else:
    f_stat = var2 / var1
    dfn, dfd = len(group2) - 1, len(group1) - 1

# -----------------------------
# 3. Compute two-tailed p-value
# -----------------------------
p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

# -----------------------------
# 4. Visualize distributions
# -----------------------------
plt.figure(figsize=(10, 6))
plt.hist(group1, bins=15, alpha=0.6, label='Group 1 (lower variance)', color='skyblue', density=True)
plt.hist(group2, bins=15, alpha=0.6, label='Group 2 (higher variance)', color='salmon', density=True)
plt.axvline(np.mean(group1), color='blue', linestyle='--', label='Mean Group 1')
plt.axvline(np.mean(group2), color='red', linestyle='--', label='Mean Group 2')
plt.title("Histogram of Two Groups with Different Variances")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# -----------------------------
# 5. Return interpretation
# -----------------------------
{
    "Group 1 Variance": round(var1, 2),
    "Group 2 Variance": round(var2, 2),
    "F-statistic": round(f_stat, 4),
    "Degrees of Freedom": (dfn, dfd),
    "P-value": round(p_value, 4),
    "Conclusion (Œ± = 0.05)": "Reject H‚ÇÄ ‚Üí Variances differ" if p_value < 0.05 else "Fail to reject H‚ÇÄ ‚Üí Variances are not significantly different"
}

```

Q27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results

```python
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# -----------------------------
# 1. Simulate observed data
# -----------------------------
# Let's simulate results from rolling a 6-sided die 120 times
np.random.seed(0)
observed = np.random.multinomial(120, [1/6]*6)
expected = np.full(6, 120 / 6)  # expected: uniform distribution

# -----------------------------
# 2. Perform Chi-square test
# -----------------------------
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# -----------------------------
# 3. Output results
# -----------------------------
print("Observed Counts :", observed)
print("Expected Counts :", expected)
print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value             : {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H‚ÇÄ ‚Üí Observed data does NOT fit expected distribution.")
else:
    print("Conclusion: Fail to reject H‚ÇÄ ‚Üí No significant difference from expected distribution.")

# -----------------------------
# 4. Optional: Plot observed vs expected
# -----------------------------
labels = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(labels))
plt.figure(figsize=(8, 5))
plt.bar(x - 0.2, observed, width=0.4, label='Observed', color='skyblue')
plt.bar(x + 0.2, expected, width=0.4, label='Expected', color='orange')
plt.xticks(x, labels)
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title('Chi-Square Goodness of Fit: Observed vs Expected')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
